```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt #import matplotlib same as from maplotlib import pyplot as plt
```

```python
#Load our external file
data = pd.read_excel('nigeria_crude_oil_prices.xlsx')
data.head()
```

```python
data.T
```

```python
data = data.head(57)
data.head()
```

```python
# Reserve the dataframe
df = data.reindex(index=data.index[::-1])
df.head()
```

```python
# Reset the dataframe
data = df.reset_index()
data.drop('index', axis=1, inplace=True)
data.head()
```

```python
# Check data type
data.dtypes
```

```python
# converting from data type object to date 
data['Date'] = pd.to_datetime(data['Date'])
data.head()
```

```python
# Extract year from datatime
data['Date'] = pd.to_datetime(data['Date']).dt.year
```
